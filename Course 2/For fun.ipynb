```python
import matplotlib.pyplot as plt
import numpy as np
from mlxtend.plotting import plot_decision_regions
```

```python
X = np.random.uniform(-4, 5, (10, 2))
y = np.zeros(X.shape[0])
```

```python
y[X[:,1] > X[:,0]] = 1
```

```python
pos = y == 1
neg = y == 0
plt.scatter(X[pos, 0], X[pos, 1], marker='x', color='red', s = 100)
plt.scatter(X[neg, 0], X[neg, 1], marker='o', color='blue', s = 100, facecolors='none')
```

```python
from sklearn.svm import SVC
model = SVC(kernel='linear')
model.fit(X, y)
```

```python
plot_decision_regions(X, y.astype(np.int), model)
```

```python
model.predict([[-1, -8]])
```

## **Vectorizing threshold or error evaluation**

```python
def get_error(y, y_hat):
    error = np.zeros(y.shape)
    error[y != y_hat] = 1
    return np.average(error)
```

```python
def get_error_2(y, y_hat):
    m = len(y)
    incorrect = 0
    for i in range(m):
        if y[i] != y_hat[i]:
            incorrect += 1
    return incorrect/m
```

```python
y = np.zeros(100000)
y_hat = np.random.randint(0, 5, 100000)
```

```python
%timeit get_error(y, y_hat)
```

```python
%timeit get_error_2(y, y_hat)
```
